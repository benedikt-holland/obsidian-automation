```python
import pandas as pd
```

```python
hist = pd.read_csv("history.csv")
hist.head()
```

```python
hist.groupby("category").count()["check"] 
```

```python
INPUT = '../../obsidian/living/fitness/Gym.md' 
with open(INPUT, 'r') as f:
    lines = f.readlines()
data = [] 
for line in lines:
    if line[0] != "|":
        continue  
    # Data is kept in a markdown table
    data.append([l.strip() for l in line[1:-2].split("|")])
header = data[0]
data = data[2:]  
df = pd.DataFrame(data, columns=header)
```

```python
orig_cols = df.columns
df.head()
```

```python
df["full_name"] = df.name + " " + df.variant
hist["full_name"] = hist.name + " " + hist.variant 
```

```python
hist.check = pd.to_datetime(hist.check)
hist["day_diff"] = hist.check.rsub(pd.Timestamp.now()).dt.days
```

```python
cols = ["category", "subcategory", "full_name"]
```

```python
group_cols = [] 
for col in cols: 
    group_cols.append(col)
    df = df.join(hist.groupby(group_cols).count().check.rename(col) , on=group_cols, rsuffix="_count")
    df[col+"_count"] = df[col+"_count"].fillna(0)
```

```python
df = df.join(hist.groupby(group_cols).min().day_diff.rename("days") , on=group_cols)
```

```python
df
```

```python
total = hist.shape[0] 
for col in [c+"_count" for c in cols]:
    df[col+"%"] = (df[col] / total).fillna(0)
    total = df[col]     
```

```python
df
```

```python
INPUT = '../../obsidian/living/fitness/Gym goals.md' 
with open(INPUT, 'r') as f:
    lines = f.readlines()
data = [] 
for line in lines:
    if line[0] != "|":
        continue  
    # Data is kept in a markdown table
    data.append([l.strip() for l in line[1:-2].split("|")])
header = data[0]
data = data[2:]  
goals = pd.DataFrame(data, columns=header)
goals
```

```python
goals.goal = goals.goal.astype(int)
```

```python
total = goals.goal.sum()
group_cols = [] 
for col in ["category", "subcategory"]:
    group_cols.append(col)
    goals = goals.join(goals.groupby(group_cols).sum()["goal"].rename(col), on=group_cols, rsuffix="_total")
    goals[col+"%"] = goals[col+"_total"] / total
    total = goals[col+"_total"] 
```

```python
goals
```

```python
df = df.join(goals.set_index(["category", "subcategory"]), on=["category", "subcategory"], rsuffix="_goal", )
df.head()
```

```python
df.columns
```

```python
for col in ["category", "subcategory"]:
    df[col+"_score"] = (df[col+"%"]  - df[col+"_count%"]) + 1
df["score"] = df.category_score + df.subcategory_score
df
```

```python
import numpy as np
```

```python
df["day_score"] = df["days"].fillna(np.inf) * df.score
df.day_score
```
